```python
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Cột cần OneHot (nhiều hơn 2 loại)
categorical_cols = ['BusinessTravel', 'Department', 'EducationField', 'JobRole']

# Cột chỉ có 2 loại → LabelEncode
binary_cols = ['MaritalStatus', 'Gender']

# Label encode binary columns
df_encoded = df.copy()
for col in binary_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])

# OneHot encode categorical columns
ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(drop='first'), categorical_cols)],
    remainder='passthrough'
)

# Áp dụng ColumnTransformer vào df_encoded
X = np.array(ct.fit_transform(df_encoded))  # Sử dụng df_encoded thay vì X
```

```python
X
```

```python
import numpy as np

X = np.delete(X, 18, axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

# Train Model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 11, metric = 'minkowski', p = 1, weights = 'uniform')
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Khởi tạo mô hình KNN
knn = KNeighborsClassifier()

# Định nghĩa các tham số cần tìm kiếm
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],  # Số lượng hàng xóm
    'metric': ['minkowski', 'euclidean', 'manhattan'],  # Chỉ số khoảng cách
    'weights': ['uniform', 'distance'],  # Trọng số
    'p': [1, 2],  # p cho khoảng cách Minkowski (1: Manhattan, 2: Euclidean)
}

# Sử dụng GridSearchCV để tìm tham số tối ưu với 'precision' làm thước đo
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='precision', n_jobs=-1)

# Huấn luyện với các tham số tốt nhất
grid_search.fit(X_train, y_train)

# In các tham số tối ưu và kết quả
print(f'Thông số tối ưu: {grid_search.best_params_}')
print(f'Precision tối ưu: {grid_search.best_score_}')
```
